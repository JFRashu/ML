# Model Training and Testing
* # Split the data
* # Convert the data from text to numeric values
* # Train the model
* # Model Evaluation

```python
X = dataset[['Outlook', 'Temperature', 'Humidity', 'Wind']]
y = dataset.Play_Badminton.values
```

```python
X
```

```python
from sklearn.preprocessing import OneHotEncoder

# One-hot encode categorical features
encoder = OneHotEncoder()
X_encoded = encoder.fit_transform(X)
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(X_encoded,y,test_size=.2 ,random_state =42)
```

```python
# cv=CountVectorizer()
x_train=xtrain
```

```python
x_train
```

```python
# model = MultinomialNB()
# model.fit(x_train,ytrain)
```

```python
model = BernoulliNB()
model.fit(x_train,ytrain)
```

# Model Performance score

```python
model.score(xtest,ytest)
```

```python
import numpy as np

# Define the new data point
new_data = [['Sunny', 'Mild', 'Normal', 'Strong']]

# Transform using the same encoder
new_data_encoded = encoder.transform(new_data)

# Convert to array
new_data_array = new_data_encoded.toarray()

# Make a prediction
prediction = model.predict(new_data_array)

# Output the result
print("Prediction:", prediction[0])  # The predicted class
```
